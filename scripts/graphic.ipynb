Графики обученных модели

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", font_scale=1.1)
```

```python
path1 = "/home/ruslana/Projects/RaspberryYolo/RaspberryRoadSign/runs/rtsd_train/rtsd_yolo11n_pi_first100/results.csv"
path2 = "/home/ruslana/Projects/RaspberryYolo/RaspberryRoadSign/runs/rtsd_train/rtsd_yolo11n_pi_other50/results.csv" 
```

```python
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)

df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("Box Loss: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='train/box_loss', label='Train', color='blue')
sns.lineplot(data=df1, x='epoch', y='val/box_loss', label='Val', color='orange')
plt.title("Run 1 (320px)")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='train/box_loss', label='Train', color='blue')
sns.lineplot(data=df2, x='epoch', y='val/box_loss', label='Val', color='orange')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("Class Loss: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='train/cls_loss', label='Train', color='blue')
sns.lineplot(data=df1, x='epoch', y='val/cls_loss', label='Val', color='orange')
plt.title("Run 1 (320px)")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='train/cls_loss', label='Train', color='blue')
sns.lineplot(data=df2, x='epoch', y='val/cls_loss', label='Val', color='orange')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("Precision: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='metrics/precision(B)', color='green')
plt.title("Run 1 (320px)")
plt.ylabel("Precision")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='metrics/precision(B)', color='green')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("Recall: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='metrics/recall(B)', color='red')
plt.title("Run 1 (320px)")
plt.ylabel("Recall")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='metrics/recall(B)', color='red')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("mAP@50: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='metrics/mAP50(B)', color='purple')
plt.title("Run 1 (320px)")
plt.ylabel("mAP@50")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='metrics/mAP50(B)', color='purple')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.suptitle("Learning Rate: Run 1 (320px) vs Run 2 (480px)", fontsize=14, weight='bold')

plt.subplot(1, 2, 1)
sns.lineplot(data=df1, x='epoch', y='lr/pg0', color='gray')
plt.title("Run 1 (320px)")
plt.ylabel("LR")

plt.subplot(1, 2, 2)
sns.lineplot(data=df2, x='epoch', y='lr/pg0', color='gray')
plt.title("Run 2 (480px)")

plt.tight_layout()
plt.show()
```
